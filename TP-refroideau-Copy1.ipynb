```python
import matplotlib.pyplot as plt
import numpy as np

#================= VALEURS EXPÉRIMENTALES MESURÉES : ==========
t_exp     =[1,2,3] #A compléter : listes de vos valeurs du temps (en s)
                   #Attention, les nombres à virgules s'écrivent avec des points
                   #Séparer chaque élément de la liste par des virgules
    
theta_exp =[4,2,3] #A compléter: listes des valeurs des températures (en °C)

T_exp = [] # Ne pas modifier : liste vide qui contiendra les valeurs des températures en degré Kelvin 

for i in range(len(theta_exp)): #A compléter
    T = theta_exp[i] + 273      # A compléter pour convertir la valeur de température en degré Kelvin
    T_exp.append(T)             # A compléter

#================= TRACÉ DU GRAPHIQUE : =========================
plt.clf()
plt.plot(t_exp,T_exp,'x')
plt.title("Valeurs expérimentales")
plt.xlabel("t (en s)")
plt.ylabel("T (en K)")

plt.show()
```

✍️**A partir du graphique ci-dessus, estimer les valeurs de la température initiale $T_{ini}$, finale $T_{fin}$ et du temps caractéristique $\tau$**

*Indice : pour trouver la valeur de $\tau$, on s'inspirera de la méthode vue dans le chapitre ELEC*
✍️**Compléter puis exécuter le code ci-dessous**

```python
T_ini   = 373       #A compléter : Température initiale (en K)
T_fin = 273         #A compléter : empérature finale   (en K)
A   = T_fin         #A compléter : Expression de la constante A de l'équation différentielle
B   = T_ini-T_fin   #A compléter : Expression de la constante B de l'équation différentielle
tau = 60            #A compléter : Valeur de tau (en s) trouvée grâce au graphique précédent
t_model = np.arange(0,5*60,10) #Ne pas modifier : liste de valeurs de t
T_model = A+B*np.exp(-t_model/tau) #Ne pas modifier : expression de la solution de l'équation différentielle

#============ TRACÉ DU GRAPHIQUE (ne pas modifier) ===========
plt.clf()
plt.plot(t_exp,T_exp,'x',label="valeurs expérimentales")
plt.plot(t_model,T_model,label="modélisation théorique : $A+B \cdot exp(-\\frac{t}{\\tau})$")
tangente = (-B/tau)*np.arange(0,tau+20,10)+T_ini
plt.plot(np.arange(0,tau+20,10),tangente,'--',label="tangente à l'origine")
plt.plot(t_model,T_fin*np.ones(len(t_model)),'--',label="asymptote en $+\\infty$")

plt.legend()
plt.grid()
plt.title("Comparaison entre la modélisation et les valeurs expérimentales")
plt.xlabel("t (en s)")
plt.ylabel("T (en K)")
axes = plt.gca()
axes.yaxis.set_ticks(np.arange(250,373,10))
axes.xaxis.set_ticks(np.arange(0,np.max([np.max(t_model),np.max(t_exp)]),40))
axes.xaxis.set_ticks(np.arange(0,np.max([np.max(t_model),np.max(t_exp)]),20), minor = True)
plt.show()
```
